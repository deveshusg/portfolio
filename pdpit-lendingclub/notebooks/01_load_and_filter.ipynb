📌 Load 1.6GB LendingClub Dataset Safely (Chunking)

```python
import pandas as pd

chunksize = 250000   # safe for 12GB RAM

dfs = []
total_rows=0
for chunk in pd.read_csv(
    RAW_PATH,
    chunksize=chunksize,
    low_memory=False,
    dtype=str  # load everything as string first
):
    # remove repeated headers that appear inside LC files
    if 'loan_status' in chunk.columns:
        chunk = chunk[chunk['loan_status'] != 'loan_status']
    dfs.append(chunk)
    total_rows += len(chunk)
    print("Rows loaded so far:", total_rows)

df_raw = pd.concat(dfs, ignore_index=True)
print("Raw dataset shape:", df_raw.shape)
df_raw.head()
```

📌 Convert Dates, Numeric Columns & Fix Types

```python
import numpy as np
from datetime import datetime

# Convert issue date (LendingClub uses "Jan-2015" style)
df_raw["issue_d"] = pd.to_datetime(df_raw["issue_d"], format="%b-%Y", errors="coerce")

# Convert key numeric columns
numeric_cols = [
    "loan_amnt", "funded_amnt", "funded_amnt_inv", "int_rate", "installment",
    "annual_inc", "dti", "delinq_2yrs", "inq_last_6mths", "open_acc",
    "total_acc", "revol_bal", "revol_util"
]

for col in numeric_cols:
    if col in df_raw.columns:
        df_raw[col] = pd.to_numeric(df_raw[col].str.replace("%",""), errors="coerce")
```

📌 Filter to PIT-Friendly Window (2015–2018)

```python
start = datetime(2015, 1, 1)
end   = datetime(2018, 12, 31)

df_filtered = df_raw[
    (df_raw["issue_d"] >= start) &
    (df_raw["issue_d"] <= end)
]

print("Filtered dataset shape:", df_filtered.shape)
df_filtered.head()
```

📌 Delete Raw File & Chunks to free up RAM

```python
del dfs
del df_raw

import gc
gc.collect()
```

📌 Select PD Modeling Columns + Create Default Flag

```python
# Expected columns for PD modelling
keep_cols = [
    "id", "issue_d", "loan_amnt", "term", "int_rate", "grade", "sub_grade",
    "emp_length", "home_ownership", "annual_inc", "verification_status",
    "purpose", "addr_state", "dti", "delinq_2yrs", "inq_last_6mths",
    "open_acc", "total_acc", "revol_bal", "revol_util", "loan_status"
]

# Drop cols not present
keep_cols = [col for col in keep_cols if col in df_filtered.columns]

df_filtered = df_filtered[keep_cols]

# Create binary default indicator
default_statuses = [
    "Charged Off", "Default", "Does not meet the credit policy. Status:Charged Off",
    "Late (31-120 days)"
]

df_filtered["default_ind"] = df_filtered["loan_status"].isin(default_statuses).astype(int)

print("Final filtered dataset shape:", df_filtered.shape)
df_filtered.head()
```

📌 Memory Check & Save Final Dataset to Parquet

```python
print("Memory used:", df_filtered.memory_usage().sum() / 1e6, "MB")

OUTFILE = OUTPUT_PATH + "lendingclub_2015_2018_filtered.parquet"
df_filtered.to_parquet(OUTFILE, index=False)

print("Saved file:", OUTFILE)
```

```python
import gzip
import shutil

# ---------------------------------------------------------------
# Path to the Parquet file we just created earlier.
# This is the file we want to compress to make it GitHub-friendly.
# ---------------------------------------------------------------
src = OUTFILE

# ---------------------------------------------------------------
# Destination path for the compressed file.
# We simply add ".gz" to the Parquet filename.
# Example: "lendingclub_2015_2018_filtered.parquet.gz"
# ---------------------------------------------------------------
dst = OUTFILE + ".gz"

# ---------------------------------------------------------------
# Open the original Parquet file in binary read mode ('rb').
# gzip requires reading raw bytes, not text.
# ---------------------------------------------------------------
with open(src, "rb") as f_in:

    # -----------------------------------------------------------
    # Open the destination file in gzip write mode ('wb').
    # This creates a compressed .gz file.
    # -----------------------------------------------------------
    with gzip.open(dst, "wb") as f_out:

        # -------------------------------------------------------
        # Copy the contents from the source file to the gzip file.
        # shutil.copyfileobj streams the data in chunks,
        # so the entire file is NEVER loaded into RAM.
        # This prevents memory crashes.
        # -------------------------------------------------------
        shutil.copyfileobj(f_in, f_out)

# ---------------------------------------------------------------
# Print a confirmation message so the user knows compression is done.
# ---------------------------------------------------------------
print("Compressed file saved:", dst)
```
